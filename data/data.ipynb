```python
import psycopg2
import random

# ============================
# 1. PostgreSQL DB 연결
# ============================
conn = psycopg2.connect(
    host="localhost",
    port=5432,
    database="septime_db",
    user="postgres",
    password="1234"
)
cursor = conn.cursor()

# ============================
# 2. 모든 clinical_id, news_score 불러오기
# ============================
cursor.execute("SELECT clinical_id, news_score FROM clinical_data;")
rows = cursor.fetchall()

# ============================
# 3. 랜덤 오프셋 계산 & 업데이트
# ============================
for clinical_id, news_score in rows:
    offset = random.randint(-2, 2)              # -2 ~ +2 정수
    new_label = news_score + offset
    new_label = max(1, min(10, new_label))      # 1~10 사이로 클리핑

    cursor.execute(
        """
        UPDATE clinical_data
        SET news_score_label = %s
        WHERE clinical_id = %s;
        """,
        (new_label, clinical_id)
    )

# ============================
# 4. 커밋 및 종료
# ============================
conn.commit()
cursor.close()
conn.close()

print("✅ news_score_label 컬럼이 랜덤하게 업데이트되었습니다.")
```

```python
from datetime import datetime, timedelta
import random
import psycopg2

conn = psycopg2.connect(
    host="localhost",
    port=5432,
    database="septime_db",
    user="postgres",
    password="1234"
)
cursor = conn.cursor()

# 환자 목록 조회
cursor.execute("SELECT DISTINCT patient_id FROM clinical_data;")
patients = [row[0] for row in cursor.fetchall()]

# 시간 설정
start_date = datetime(2025, 1, 1, 0, 0, 0)
end_date = datetime(2025, 1, 4, 23, 59, 59)
latest_start_date = end_date - timedelta(hours=(10 - 1) * 8)


for pid in patients:
    # 환자별 시작 시간을 랜덤으로 뽑되, 마지막 timestamp가 end_date를 넘지 않도록 함
    random_start = start_date + timedelta(
        seconds=random.randint(0, int((latest_start_date - start_date).total_seconds()))
    )

    cursor.execute("""
        SELECT clinical_id, timepoint
        FROM clinical_data
        WHERE patient_id = %s
        ORDER BY timepoint;
    """, (pid,))
    records = cursor.fetchall()

    for clinical_id, timepoint in records:
        ts = random_start + timedelta(hours=(timepoint - 1) * 8)

        cursor.execute("""
            UPDATE clinical_data
            SET timestamp = %s
            WHERE clinical_id = %s;
        """, (ts, clinical_id))

conn.commit()
cursor.close()
conn.close()

print("✅ 모든 환자 timestamp가 2025-01-01~2025-01-03 사이에서 8시간 간격으로 정확히 설정되었습니다.")
```

```python
import psycopg2

# ============================
# 1. DB 연결
# ============================
conn = psycopg2.connect(
    host="localhost",
    port=5432,
    database="septime_db",
    user="postgres",
    password="1234"
)
cursor = conn.cursor()

# ============================
# 2. 컬럼 추가 (이미 추가했으면 생략 가능)
# ============================
cursor.execute("""
ALTER TABLE patient
ADD COLUMN IF NOT EXISTS doctor_name VARCHAR(100),
ADD COLUMN IF NOT EXISTS hospital_name VARCHAR(200);
""")

# ============================
# 3. 더미 데이터 정의
# ============================
dummy_data = [
    (1, "박철수", "SKALA대학병원"),
    (2, "김지훈", "SKALA대학병원"),
    (3, "이영희", "SKALA대학병원"),
    (4, "정우성", "SKALA대학병원"),
    (5, "한지민", "SKALA대학병원"),
    (6, "최민호", "SKALA대학병원"),
    (7, "오하늘", "SKALA대학병원"),
    (8, "김도현", "SKALA대학병원"),
    (9, "송혜교", "SKALA대학병원"),
    (10, "김범수", "SKALA대학병원")
]

# ============================
# 4. 업데이트 실행
# ============================
for pid, doctor, hospital in dummy_data:
    cursor.execute(
        """
        UPDATE patient
        SET doctor_name = %s, hospital_name = %s
        WHERE patient_id = %s;
        """,
        (doctor, hospital, pid)
    )

conn.commit()
cursor.close()
conn.close()

print("✅ patient 테이블에 담당의 / 병원 더미 데이터 추가 완료")
```
